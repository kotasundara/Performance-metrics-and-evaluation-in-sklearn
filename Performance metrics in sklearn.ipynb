```python
# Load libraries
from sklearn.metrics import make_scorer, r2_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge
from sklearn.datasets import make_regression
```

```python
# Generate features matrix and target vector
X, y = make_regression(n_samples = 100,
                          n_features = 3,
                          random_state = 1)
```

```python
X
```

```python
y.shape
```

```python
# Create training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.10, random_state=1)
```

```python
# Train model
ridge=Ridge()
model=ridge.fit(X_train,y_train)
```

```python
# Create custom metric
def custom_metric(y_test, y_pred):
    # Calculate r-squared score
    r2 = r2_score(y_test, y_pred)
    # Return r-squared score
    return r2
```

```python
# Make scorer and define that higher scores are better
score2 = make_scorer(custom_metric, greater_is_better=True)
```

```python
score2(model,X_test,y_test)
```

**# Plotting ROC(receiver operating characteristic) curve*****

```python
# Load libraries
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_curve, roc_auc_score
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
# Create feature matrix and target vector
X, y = make_classification(n_samples=10000, 
                           n_features=10, 
                           n_classes=2, 
                           n_informative=3,
                           random_state=3)
```

```python
# Split into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=1)
```

```python
# Create classifier
clf = LogisticRegression()

# Train model
clf.fit(X_train, y_train)
```

```python
# Get predicted probabilities
probability=clf.predict_proba(X_test)
```

```python
y_pred=probability[:,1]
```

```python
# Plot ROC curve
fpr,tpr,thr=roc_curve(y_test,y_pred)
```

```python
plt.title("ROC curve")
plt.plot(fpr,tpr)
plt.plot([0,1],[0,1],ls="--")
plt.vlines(x=0,ymin=0,ymax=1,colors="black")
plt.hlines(y=1,xmin=0,xmax=1,colors="black")
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```

```python
# Load libraries
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
# Load the digits dataset
digits = datasets.load_digits()

# Create the features matrix
X = digits.data

# Create the target vector
y = digits.target
```

```python
# Create training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y, 
                                                    test_size=0.1, 
                                                    random_state=1)
```

```python
# Create standardizer
sc = StandardScaler()

# Fit standardizer to training set
sc.fit(X_train)
```

```python
# Apply to both training and test sets
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)
```
